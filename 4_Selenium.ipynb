```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
# https://search.naver.com/search.naver?where=view&sm=tab_jum&query=대학축제
# https://search.naver.com/search.naver?where=view&sm=tab_jum&query=아카라카
```

```python
# get_soup 함수 만들기 (입력받은 URL의 HTML 소스코드를 가져와서 BeautifulSoup 객체로 파싱하여 반환) + 검색어를 지정해서

def get_soup(search_word):

    url="https://search.naver.com/search.naver?where=view&sm=tab_jum&query="+search_word

    res = requests.get(url)
    print(res) #

    time.sleep(1) #time.sleep은 1초 동안 코드 실행 지연하는 것

    soup = BeautifulSoup(res.text, 'html.parser')

    print(url) # 확인차원

    return soup

soup=get_soup("대학축제")
```
